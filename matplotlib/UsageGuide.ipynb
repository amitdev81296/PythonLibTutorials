# Typically one finds oneself making the same plots over and over again, but with different datasets. The recommended plotting function signature is given below

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph
    
    Parameters
    1. ax: Axes to draw to
    2. data1: The x data
    3. data2: The y data
    4, param_dict: Dictionary of kwargs to pass to ax.plot
    
    Returns
    out: List of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
```

```python
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'o'})
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data2, data3, {'marker': 'o'})
my_plotter(ax3, data3, data4, {'marker': '.'})
my_plotter(ax4, data1, data4, {'marker': 'v'})
```
